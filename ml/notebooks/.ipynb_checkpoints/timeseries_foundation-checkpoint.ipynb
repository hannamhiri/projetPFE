# Modèles de Séries Temporelles — Classiques + Foundation Models
## SARIMA | Holt-Winters | Chronos | TimesFM
## Données : Jan 2022 → Déc 2025 | Prévision : Jan 2026 → Déc 2026
### ⚠️ Métriques : MAE, RMSE, MAPE, SMAPE, Biais (pas R²)

## 1. Installation des packages

```python
# Installer les packages nécessaires
import subprocess

packages = [
    'statsmodels',
    'chronos-forecasting',   # Amazon Chronos
    'timesfm',               # Google TimesFM
]

for pkg in packages:
    print(f'Installing {pkg}...')
    result = subprocess.run(
        ['pip', 'install', pkg, '--quiet'],
        capture_output=True, text=True
    )
    if result.returncode == 0:
        print(f'  ✅ {pkg} installé')
    else:
        print(f'  ⚠️  {pkg} : {result.stderr[:100]}')

print('\n✅ Installation terminée')
```

## 2. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import clickhouse_connect
import warnings
import torch

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_error, mean_squared_error

warnings.filterwarnings('ignore')
pd.set_option('display.float_format', '{:,.2f}'.format)
plt.rcParams['figure.figsize'] = (14, 5)
sns.set_theme(style='whitegrid')

print('✅ Imports de base OK')
print(f'   PyTorch : {torch.__version__}')
print(f'   CUDA    : {torch.cuda.is_available()}')
```

## 3. Chargement des données depuis ClickHouse

```python
client = clickhouse_connect.get_client(
    host='clickhouse', port=8123,
    user='default', password='clickhouse123',
    database='gold'
)

df = client.query_df("""
SELECT
    d.Year  AS year,
    d.Month AS month,
    SUM(f.LinePrice) AS total_sales
FROM gold.fact_sales AS f
JOIN gold.dim_date AS d ON f.DateSK = d.DateSK
WHERE f.DocumentTypeSK IN (4, 1, 5)
  AND f.LinePrice IS NOT NULL
GROUP BY year, month
ORDER BY year, month
""")

df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# Nettoyage Jan 2022 → Déc 2025
df = df[
    (df['date'] >= '2022-01-01') & (df['date'] < '2026-01-01')
].reset_index(drop=True)

# Série temporelle indexée
ts = df.set_index('date')['total_sales']
ts.index.freq = 'MS'

print(f'✅ {len(ts)} mois — {ts.index[0].strftime("%b %Y")} → {ts.index[-1].strftime("%b %Y")}')
print(f'   CA moyen : {ts.mean():,.0f} DT')
print(f'   CA min   : {ts.min():,.0f} DT')
print(f'   CA max   : {ts.max():,.0f} DT')

# Visualiser
fig, ax = plt.subplots()
ax.plot(ts.index, ts.values, color='steelblue', marker='o', linewidth=2)
ax.axhline(ts.mean(), color='red', linestyle='--', label=f'Moyenne : {ts.mean():,.0f} DT')
ax.set_title('📈 CA Mensuel Jan 2022 → Déc 2025')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
plt.tight_layout()
plt.show()
```

## 4. Split Train / Test

```python
N_TEST   = 6
ts_train = ts.iloc[:-N_TEST]
ts_test  = ts.iloc[-N_TEST:]

print(f'Train : {len(ts_train)} mois ({ts_train.index[0].strftime("%b %Y")} → {ts_train.index[-1].strftime("%b %Y")})')
print(f'Test  : {len(ts_test)} mois ({ts_test.index[0].strftime("%b %Y")} → {ts_test.index[-1].strftime("%b %Y")})')

fig, ax = plt.subplots()
ax.plot(ts_train.index, ts_train.values, color='steelblue', marker='o', label='Train')
ax.plot(ts_test.index,  ts_test.values,  color='orange',    marker='o', label='Test')
ax.set_title('📊 Split Train / Test')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
plt.tight_layout()
plt.show()
```

## 5. Fonction d'évaluation — sans R²

```python
def evaluate_ts(name, y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    smape = np.mean(2 * np.abs(y_true - y_pred) /
                   (np.abs(y_true) + np.abs(y_pred))) * 100
    bias  = np.mean(y_pred - y_true)
    return {
        'name': name, 'pred': y_pred,
        'MAE': mae, 'RMSE': rmse,
        'MAPE': mape, 'SMAPE': smape, 'Biais': bias
    }

all_results = []
print('✅ Fonction evaluate_ts définie (sans R²)')
```

## 6. Modèle 1 — Holt-Winters Additif

```python
print('='*50)
print('1️⃣  Holt-Winters Additif')
print('='*50)

hw_add = ExponentialSmoothing(
    ts_train, trend='add', seasonal='add', seasonal_periods=12
).fit(optimized=True)

pred_hw_add = hw_add.forecast(N_TEST)
res_hw_add  = evaluate_ts('HW-Additif', ts_test.values, pred_hw_add.values)
all_results.append(res_hw_add)

print(f'MAE   : {res_hw_add["MAE"]:>12,.0f} DT')
print(f'RMSE  : {res_hw_add["RMSE"]:>12,.0f} DT')
print(f'MAPE  : {res_hw_add["MAPE"]:>12.2f} %')
print(f'SMAPE : {res_hw_add["SMAPE"]:>12.2f} %')
print(f'Biais : {res_hw_add["Biais"]:>12,.0f} DT')
```

## 7. Modèle 2 — Holt-Winters Multiplicatif

```python
print('='*50)
print('2️⃣  Holt-Winters Multiplicatif')
print('='*50)

hw_mul = ExponentialSmoothing(
    ts_train, trend='add', seasonal='mul', seasonal_periods=12
).fit(optimized=True)

pred_hw_mul = hw_mul.forecast(N_TEST)
res_hw_mul  = evaluate_ts('HW-Multiplicatif', ts_test.values, pred_hw_mul.values)
all_results.append(res_hw_mul)

print(f'MAE   : {res_hw_mul["MAE"]:>12,.0f} DT')
print(f'RMSE  : {res_hw_mul["RMSE"]:>12,.0f} DT')
print(f'MAPE  : {res_hw_mul["MAPE"]:>12.2f} %')
print(f'SMAPE : {res_hw_mul["SMAPE"]:>12.2f} %')
print(f'Biais : {res_hw_mul["Biais"]:>12,.0f} DT')
```

## 8. Modèle 3 — SARIMA (Grid Search)

```python
import itertools

print('='*50)
print('3️⃣  SARIMA — Grid Search')
print('='*50)

best_aic   = np.inf
best_order = None
best_sarima = None

for p, d, q in itertools.product([0,1,2], [0,1], [0,1,2]):
    try:
        m = SARIMAX(
            ts_train,
            order=(p, d, q),
            seasonal_order=(1, 1, 1, 12),
            enforce_stationarity=False,
            enforce_invertibility=False
        ).fit(disp=False)
        if m.aic < best_aic:
            best_aic    = m.aic
            best_order  = (p, d, q)
            best_sarima = m
    except:
        continue

print(f'✅ Meilleur ordre : SARIMA{best_order}(1,1,1,12) | AIC={best_aic:.2f}')

pred_sarima = best_sarima.forecast(steps=N_TEST)
res_sarima  = evaluate_ts(f'SARIMA{best_order}', ts_test.values, pred_sarima.values)
all_results.append(res_sarima)

print(f'MAE   : {res_sarima["MAE"]:>12,.0f} DT')
print(f'RMSE  : {res_sarima["RMSE"]:>12,.0f} DT')
print(f'MAPE  : {res_sarima["MAPE"]:>12.2f} %')
print(f'SMAPE : {res_sarima["SMAPE"]:>12.2f} %')
print(f'Biais : {res_sarima["Biais"]:>12,.0f} DT')
```

## 9. Modèle 4 — Amazon Chronos
### Foundation Model pré-entraîné sur des millions de séries temporelles

```python
print('='*50)
print('4️⃣  Amazon Chronos')
print('='*50)

try:
    from chronos import ChronosPipeline

    # Charger le modèle Chronos tiny (plus léger)
    print('Chargement du modèle Chronos...')
    pipeline = ChronosPipeline.from_pretrained(
        'amazon/chronos-t5-tiny',  # tiny=léger, small/base/large=plus précis
        device_map='cpu',
        torch_dtype=torch.float32,
    )
    print('✅ Chronos chargé')

    # Préparer la série
    context = torch.tensor(ts_train.values, dtype=torch.float32).unsqueeze(0)

    # Prédiction avec intervalles de confiance
    forecast = pipeline.predict(
        context=context,
        prediction_length=N_TEST,
        num_samples=100,
    )

    # Médiane des prédictions
    pred_chronos = np.median(forecast[0].numpy(), axis=0)
    low_chronos  = np.percentile(forecast[0].numpy(), 10, axis=0)
    high_chronos = np.percentile(forecast[0].numpy(), 90, axis=0)

    res_chronos = evaluate_ts('Chronos-tiny', ts_test.values, pred_chronos)
    all_results.append(res_chronos)

    print(f'MAE   : {res_chronos["MAE"]:>12,.0f} DT')
    print(f'RMSE  : {res_chronos["RMSE"]:>12,.0f} DT')
    print(f'MAPE  : {res_chronos["MAPE"]:>12.2f} %')
    print(f'SMAPE : {res_chronos["SMAPE"]:>12.2f} %')
    print(f'Biais : {res_chronos["Biais"]:>12,.0f} DT')
    CHRONOS_OK = True

except Exception as e:
    print(f'⚠️  Chronos non disponible : {e}')
    print('   → Continuer avec les autres modèles')
    CHRONOS_OK = False
    pred_chronos = None
    low_chronos  = None
    high_chronos = None
```

## 10. Modèle 5 — Google TimesFM
### Foundation Model de Google pré-entraîné sur 100 milliards de points

```python
print('='*50)
print('5️⃣  Google TimesFM')
print('='*50)

try:
    import timesfm

    print('Chargement du modèle TimesFM...')
    tfm = timesfm.TimesFm(
        hparams=timesfm.TimesFmHparams(
            backend='cpu',
            per_core_batch_size=32,
            horizon_len=N_TEST,
        ),
        checkpoint=timesfm.TimesFmCheckpoint(
            huggingface_repo_id='google/timesfm-1.0-200m-pytorch'
        ),
    )
    print('✅ TimesFM chargé')

    # Prédiction
    forecast_input = [ts_train.values.tolist()]
    frequency_input = [0]  # 0=unknown, 1=secondly, ...

    point_forecast, experimental_quantile_forecast = tfm.forecast(
        forecast_input,
        freq=frequency_input,
    )

    pred_timesfm = point_forecast[0][-N_TEST:]
    res_timesfm  = evaluate_ts('TimesFM', ts_test.values, pred_timesfm)
    all_results.append(res_timesfm)

    print(f'MAE   : {res_timesfm["MAE"]:>12,.0f} DT')
    print(f'RMSE  : {res_timesfm["RMSE"]:>12,.0f} DT')
    print(f'MAPE  : {res_timesfm["MAPE"]:>12.2f} %')
    print(f'SMAPE : {res_timesfm["SMAPE"]:>12.2f} %')
    print(f'Biais : {res_timesfm["Biais"]:>12,.0f} DT')
    TIMESFM_OK = True

except Exception as e:
    print(f'⚠️  TimesFM non disponible : {e}')
    print('   → Continuer avec les autres modèles')
    TIMESFM_OK = False
    pred_timesfm = None
```

## 11. Comparaison de tous les modèles

```python
df_results = pd.DataFrame([
    {k: v for k, v in r.items() if k != 'pred'}
    for r in all_results
]).set_index('name')

df_results_sorted = df_results.sort_values('MAPE')

print('=' * 75)
print(f'{"Modèle":<22} {"MAE":>10} {"RMSE":>10} {"MAPE":>8} {"SMAPE":>8} {"Biais":>12}')
print('=' * 75)
for _, row in df_results_sorted.iterrows():
    print(f'{row.name:<22} {row["MAE"]:>10,.0f} {row["RMSE"]:>10,.0f} '
          f'{row["MAPE"]:>7.2f}% {row["SMAPE"]:>7.2f}% {row["Biais"]:>12,.0f}')
print('=' * 75)

best_ts_name = df_results['MAPE'].idxmin()
print(f'\n🏆 Meilleur modèle : {best_ts_name} (MAPE={df_results.loc[best_ts_name, "MAPE"]:.2f}%)')

# Graphe comparaison MAPE
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

colors = plt.cm.Set2.colors
for ax, metric in zip(axes, ['MAPE', 'MAE']):
    vals = df_results_sorted[metric]
    bars = ax.bar(vals.index, vals.values,
                  color=[colors[i % len(colors)] for i in range(len(vals))])
    ax.set_title(f'📊 {metric} par Modèle')
    ax.set_ylabel(metric)
    ax.tick_params(axis='x', rotation=30)
    for bar, val in zip(bars, vals.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + vals.max()*0.01,
                f'{val:.1f}%' if metric == 'MAPE' else f'{val:,.0f}',
                ha='center', fontsize=8)

plt.suptitle('Comparaison Modèles Séries Temporelles', fontsize=13)
plt.tight_layout()
plt.show()
```

## 12. Visualisation CA Réel vs Prédit — tous les modèles

```python
fig, ax = plt.subplots(figsize=(14, 6))

# Train
ax.plot(ts_train.index, ts_train.values,
        color='steelblue', linewidth=1.5, alpha=0.6, label='Train')

# Test réel
ax.plot(ts_test.index, ts_test.values,
        color='black', marker='o', linewidth=2.5, label='CA Réel (test)', zorder=5)

# Prédictions
palette = plt.cm.tab10.colors
for i, r in enumerate(all_results):
    ax.plot(ts_test.index, r['pred'],
            label=f"{r['name']} (MAPE={r['MAPE']:.1f}%)",
            marker='s', linestyle='--',
            color=palette[i % len(palette)], alpha=0.8)

# Intervalles de confiance Chronos
if CHRONOS_OK and low_chronos is not None:
    ax.fill_between(ts_test.index, low_chronos, high_chronos,
                    alpha=0.15, color='orange', label='Chronos IC 80%')

ax.set_title('📈 CA Réel vs Prédit — Comparaison Modèles Séries Temporelles')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend(loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

## 13. Prédiction 2026 — Meilleur modèle

```python
print(f'Réentraînement de {best_ts_name} sur toute la série ({len(ts)} mois)...')

N_PRED = 12  # Jan 2026 → Déc 2026

if 'HW-Additif' in best_ts_name:
    final_model  = ExponentialSmoothing(
        ts, trend='add', seasonal='add', seasonal_periods=12
    ).fit(optimized=True)
    forecast_vals = final_model.forecast(N_PRED).values

elif 'HW-Multiplicatif' in best_ts_name:
    final_model  = ExponentialSmoothing(
        ts, trend='add', seasonal='mul', seasonal_periods=12
    ).fit(optimized=True)
    forecast_vals = final_model.forecast(N_PRED).values

elif 'SARIMA' in best_ts_name:
    final_model = SARIMAX(
        ts, order=best_order, seasonal_order=(1, 1, 1, 12),
        enforce_stationarity=False, enforce_invertibility=False
    ).fit(disp=False)
    forecast_vals = final_model.forecast(steps=N_PRED).values

elif 'Chronos' in best_ts_name and CHRONOS_OK:
    context = torch.tensor(ts.values, dtype=torch.float32).unsqueeze(0)
    forecast = pipeline.predict(context=context, prediction_length=N_PRED, num_samples=100)
    forecast_vals = np.median(forecast[0].numpy(), axis=0)

elif 'TimesFM' in best_ts_name and TIMESFM_OK:
    forecast_input = [ts.values.tolist()]
    point_forecast, _ = tfm.forecast(forecast_input, freq=[0])
    forecast_vals = point_forecast[0][-N_PRED:]

# Dates futures Jan 2026 → Déc 2026
future_dates = pd.date_range(
    start=ts.index[-1] + pd.DateOffset(months=1),
    periods=N_PRED, freq='MS'
)

df_pred = pd.DataFrame({
    'date':            future_dates,
    'year':            future_dates.year,
    'month':           future_dates.month,
    'predicted_sales': forecast_vals,
    'is_prediction':   1
})

print(f'\n✅ Prédictions {best_ts_name} :')
for _, row in df_pred.iterrows():
    print(f'  {row["date"].strftime("%b %Y")} → {row["predicted_sales"]:>15,.2f} DT')

print(f'\nCA prévu 2026 : {df_pred["predicted_sales"].sum():,.2f} DT')
print(f'CA réel  2025 : {ts[ts.index.year == 2025].sum():,.2f} DT')
```

## 14. Visualisation finale — Réel + Prédit 2026

```python
import plotly.graph_objects as go

fig = go.Figure()

# CA réel
fig.add_trace(go.Scatter(
    x=ts.index, y=ts.values,
    name='CA Réel', mode='lines+markers',
    line=dict(color='steelblue', width=2), marker=dict(size=5)
))

# CA prédit
fig.add_trace(go.Scatter(
    x=df_pred['date'], y=df_pred['predicted_sales'],
    name=f'CA Prédit 2026 ({best_ts_name})',
    mode='lines+markers',
    line=dict(color='orange', width=2, dash='dash'),
    marker=dict(size=6, symbol='square')
))

# Intervalles de confiance si Chronos
if 'Chronos' in best_ts_name and CHRONOS_OK:
    fig.add_trace(go.Scatter(
        x=list(df_pred['date']) + list(df_pred['date'])[::-1],
        y=list(high_chronos) + list(low_chronos)[::-1],
        fill='toself', fillcolor='rgba(255,165,0,0.2)',
        line=dict(color='rgba(255,165,0,0)'),
        name='IC 80%'
    ))

fig.add_vline(
    x=ts.index[-1].timestamp() * 1000,
    line_dash='dot', line_color='red',
    annotation_text='Déc 2025 — Fin données réelles'
)

fig.update_layout(
    title=f'📈 CA Réel (Jan 2022→Déc 2025) + Prédit (Jan→Déc 2026) — {best_ts_name}',
    xaxis_title='Mois', yaxis_title='CA (DT)',
    hovermode='x unified', height=500
)
fig.show()
```

## 15. CA annuel 2026 prévu

```python
ca_2024 = ts[ts.index.year == 2024].sum()
ca_2025 = ts[ts.index.year == 2025].sum()
ca_2026 = df_pred['predicted_sales'].sum()
croissance = (ca_2026 - ca_2025) / ca_2025 * 100

print('=' * 55)
print('📋 PRÉVISION CA ANNUEL 2026')
print('=' * 55)
print(f'CA réel 2024       : {ca_2024:>15,.2f} DT')
print(f'CA réel 2025       : {ca_2025:>15,.2f} DT')
print(f'CA prévu 2026      : {ca_2026:>15,.2f} DT')
print(f'Croissance 25→26   : {croissance:>15.2f} %')
print(f'Modèle             : {best_ts_name}')
print(f'MAPE test          : {df_results.loc[best_ts_name, "MAPE"]:.2f}%')
print('=' * 55)

# Graphe
years_data = pd.DataFrame([
    {'Année': '2024 (réel)',  'CA': ca_2024},
    {'Année': '2025 (réel)',  'CA': ca_2025},
    {'Année': '2026 (prédit)','CA': ca_2026},
])
fig, ax = plt.subplots(figsize=(8, 5))
colors  = ['steelblue', 'steelblue', 'orange']
ax.bar(years_data['Année'], years_data['CA'], color=colors)
for i, (_, r) in enumerate(years_data.iterrows()):
    ax.text(i, r['CA'] + years_data['CA'].max() * 0.01,
            f"{r['CA']:,.0f} DT", ha='center', fontsize=9)
ax.set_title('📊 CA Annuel : 2024 Réel | 2025 Réel | 2026 Prévu')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.tight_layout()
plt.show()
```
