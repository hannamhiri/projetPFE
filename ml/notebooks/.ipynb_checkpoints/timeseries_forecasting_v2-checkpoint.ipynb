## 1. Chargement des données

```python
client = clickhouse_connect.get_client(
    host='clickhouse', port=8123,
    user='default', password='clickhouse123',
    database='gold'
)

QUERY = """
SELECT
    toYear(toDate(toString(fs.DateSK), '%Y%m%d'))  AS year,
    toMonth(toDate(toString(fs.DateSK), '%Y%m%d')) AS month,
    sum(fs.LinePrice)                               AS total_sales
FROM gold.fact_sales fs
JOIN gold.dim_document_type    ddt ON fs.DocumentTypeSK   = ddt.DocumentTypeSK
JOIN gold.dim_document_status  dds ON fs.DocumentStatusSK = dds.DocumentStatusSK
WHERE fs.DocumentTypeSK   IN (1, 4, 6)
  AND fs.DocumentStatusSK NOT IN (4, 13)
  AND fs.LinePrice > 0
  AND fs.DateSK IS NOT NULL
GROUP BY year, month
ORDER BY year, month
"""

result = client.query(QUERY)
df = pd.DataFrame(result.result_rows, columns=result.column_names)
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
df = df.sort_values('date').reset_index(drop=True)

# Série temporelle indexée
ts = df.set_index('date')['total_sales']
ts.index = pd.DatetimeIndex(ts.index, freq='MS')

print(f"✅ {len(ts)} mois chargés")
print(f"   Période  : {ts.index[0].strftime('%b %Y')} → {ts.index[-1].strftime('%b %Y')}")
print(f"   CA moyen : {ts.mean():>15,.2f} DT")
print(f"   CA min   : {ts.min():>15,.2f} DT")
print(f"   CA max   : {ts.max():>15,.2f} DT")
print(f"   CV       : {ts.std()/ts.mean():.2%}  (< 30% = régulier, > 50% = chaotique)")
df.tail(6)
```

## 3. Test de Stationnarité (ADF) & ACF/PACF

```python
# Test ADF
adf_result = adfuller(ts.dropna())
print("=" * 50)
print("Test ADF (Augmented Dickey-Fuller)")
print("=" * 50)
print(f"  Statistique ADF : {adf_result[0]:.4f}")
print(f"  p-value         : {adf_result[1]:.4f}")
print(f"  Seuils critiques:")
for key, val in adf_result[4].items():
    print(f"    {key}: {val:.4f}")
if adf_result[1] < 0.05:
    print("\n✅ Série STATIONNAIRE (p < 0.05) → pas de différenciation nécessaire")
else:
    print("\n⚠️  Série NON STATIONNAIRE (p > 0.05) → d=1 recommandé pour SARIMA")

# ACF / PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(ts.dropna(),  lags=24, ax=axes[0], title='ACF — Autocorrélation')
plot_pacf(ts.dropna(), lags=24, ax=axes[1], title='PACF — Autocorrélation Partielle')
plt.tight_layout()
plt.show()
```

## 4. Walk-Forward Validation
Contrairement à un simple split train/test, la **walk-forward validation** simule les vraies conditions de prévision :
à chaque étape, on entraîne sur tout l'historique disponible et on prédit le mois suivant.
Cela donne une estimation **fiable et non biaisée** des performances.


```python
def walk_forward_validation(ts, model_fn, n_test=12):
    """
    Walk-forward validation sur n_test mois.
    model_fn(train) → array de prévisions 1 mois
    Retourne : actuals, predictions, métriques
    """
    n = len(ts)
    n_train_start = n - n_test
    actuals = []
    preds   = []

    for i in range(n_test):
        train = ts.iloc[:n_train_start + i]
        actual = ts.iloc[n_train_start + i]
        try:
            pred = model_fn(train)
            preds.append(float(pred))
        except Exception as e:
            preds.append(np.nan)
        actuals.append(float(actual))

    actuals = np.array(actuals)
    preds   = np.array(preds)
    mask    = ~np.isnan(preds)

    mape = np.mean(np.abs((actuals[mask] - preds[mask]) / actuals[mask])) * 100
    mae  = np.mean(np.abs(actuals[mask] - preds[mask]))
    rmse = np.sqrt(np.mean((actuals[mask] - preds[mask])**2))
    bias = np.mean(preds[mask] - actuals[mask])

    return actuals, preds, {
        'MAPE': round(mape, 2),
        'MAE':  round(mae, 0),
        'RMSE': round(rmse, 0),
        'Biais': round(bias, 0),
    }

print("✅ Fonction walk-forward définie")
print(f"   Test sur les {12} derniers mois")
print(f"   Train initial : {len(ts) - 12} mois")
```

## 5. Holt-Winters — Grid Search optimisé

```python
print("⏳ Grid search Holt-Winters (toutes combinaisons trend/seasonal)...\n")

hw_configs = [
    {'trend': 'add',  'seasonal': 'add',  'label': 'HW Additif'},
    {'trend': 'add',  'seasonal': 'mul',  'label': 'HW Multiplicatif'},
    {'trend': 'mul',  'seasonal': 'add',  'label': 'HW Trend-Mul / Seas-Add'},
    {'trend': None,   'seasonal': 'add',  'label': 'HW Sans Trend (Add)'},
    {'trend': None,   'seasonal': 'mul',  'label': 'HW Sans Trend (Mul)'},
]

hw_results = {}

for cfg in hw_configs:
    def make_hw_fn(trend, seasonal):
        def hw_fn(train):
            model = ExponentialSmoothing(
                train,
                trend=trend,
                seasonal=seasonal,
                seasonal_periods=12,
                initialization_method='estimated',
            ).fit(optimized=True)
            return model.forecast(1).iloc[0]
        return hw_fn

    actuals, preds, metrics = walk_forward_validation(
        ts, make_hw_fn(cfg['trend'], cfg['seasonal']), n_test=12
    )
    hw_results[cfg['label']] = {
        'actuals': actuals, 'preds': preds, **metrics
    }
    print(f"  {cfg['label']:<35} MAPE={metrics['MAPE']:>6.2f}%  MAE={metrics['MAE']:>10,.0f}  RMSE={metrics['RMSE']:>10,.0f}")

# Meilleur HW
best_hw_name = min(hw_results, key=lambda k: hw_results[k]['MAPE'])
best_hw      = hw_results[best_hw_name]
print(f"\n🏆 Meilleur Holt-Winters : {best_hw_name}  →  MAPE = {best_hw['MAPE']:.2f}%")
```

## 6. SARIMA — Grid Search automatique

```python
print("⏳ Grid search SARIMA (p,d,q)(P,D,Q,12)...\n")

# Grille raisonnable
p_vals = [0, 1, 2]
d_vals = [1]
q_vals = [0, 1, 2]
P_vals = [0, 1]
D_vals = [1]
Q_vals = [0, 1]

best_sarima_mape = np.inf
best_sarima_order = None
best_sarima_seasonal = None
sarima_results_all = {}

for p, d, q, P, D, Q in product(p_vals, d_vals, q_vals, P_vals, D_vals, Q_vals):
    order          = (p, d, q)
    seasonal_order = (P, D, Q, 12)
    label = f"SARIMA{order}x{seasonal_order[:3]}"

    def make_sarima_fn(order, seasonal_order):
        def sarima_fn(train):
            model = SARIMAX(
                train,
                order=order,
                seasonal_order=seasonal_order,
                enforce_stationarity=False,
                enforce_invertibility=False,
            ).fit(disp=False)
            return model.forecast(1).iloc[0]
        return sarima_fn

    try:
        actuals, preds, metrics = walk_forward_validation(
            ts, make_sarima_fn(order, seasonal_order), n_test=12
        )
        sarima_results_all[label] = {'actuals': actuals, 'preds': preds, **metrics}
        if metrics['MAPE'] < best_sarima_mape:
            best_sarima_mape     = metrics['MAPE']
            best_sarima_order    = order
            best_sarima_seasonal = seasonal_order
            best_sarima_label    = label
    except:
        pass

print(f"🏆 Meilleur SARIMA : {best_sarima_label}")
print(f"   Order    : {best_sarima_order}")
print(f"   Seasonal : {best_sarima_seasonal}")
print(f"   MAPE     : {best_sarima_mape:.2f}%")

best_sarima = sarima_results_all[best_sarima_label]
```

## 7. Prophet — Calibré pour données B2B mensuelles

```python
print("⏳ Walk-forward validation Prophet...\n")

def prophet_fn(train):
    df_p = pd.DataFrame({
        'ds': train.index,
        'y':  train.values
    })
    model = Prophet(
        seasonality_mode          = 'multiplicative',
        yearly_seasonality        = True,
        weekly_seasonality        = False,
        daily_seasonality         = False,
        changepoint_prior_scale   = 0.05,   # regularisation tendance
        seasonality_prior_scale   = 10.0,
        interval_width            = 0.80,
        changepoint_range         = 0.9,
    )
    model.add_seasonality(name='quarterly', period=91.25, fourier_order=5)
    model.fit(df_p)
    future = model.make_future_dataframe(periods=1, freq='MS')
    forecast = model.predict(future)
    return forecast.iloc[-1]['yhat']

actuals_p, preds_p, metrics_p = walk_forward_validation(ts, prophet_fn, n_test=12)
prophet_results = {'actuals': actuals_p, 'preds': preds_p, **metrics_p}

print(f"✅ Prophet  MAPE={metrics_p['MAPE']:.2f}%  MAE={metrics_p['MAE']:,.0f}  RMSE={metrics_p['RMSE']:,.0f}")
```

## 8. Comparaison des Modèles

```python
all_results = {}

# Meilleur HW
all_results[best_hw_name] = best_hw

# Top 3 SARIMA
top_sarima = sorted(sarima_results_all.items(), key=lambda x: x[1]['MAPE'])[:3]
for name, res in top_sarima:
    all_results[name] = res

# Prophet
all_results['Prophet'] = prophet_results

# Tableau
rows = []
for name, res in all_results.items():
    rows.append({
        'Modèle': name,
        'MAPE (%)': res['MAPE'],
        'MAE': res['MAE'],
        'RMSE': res['RMSE'],
        'Biais': res['Biais'],
    })

df_results = pd.DataFrame(rows).sort_values('MAPE (%)').reset_index(drop=True)

print("=" * 85)
print(f"{'Modèle':<40} {'MAPE':>8} {'MAE':>12} {'RMSE':>12} {'Biais':>12}")
print("=" * 85)
for _, row in df_results.iterrows():
    flag = " ← 🏆" if _ == 0 else ""
    print(f"  {row['Modèle']:<38} {row['MAPE (%)']:>7.2f}%  {row['MAE']:>12,.0f}  {row['RMSE']:>12,.0f}  {row['Biais']:>12,.0f}{flag}")
print("=" * 85)

best_name    = df_results.iloc[0]['Modèle']
best_results = all_results[best_name]
print(f"\n🏆 Meilleur modèle global : {best_name}  →  MAPE = {df_results.iloc[0]['MAPE (%)']:.2f}%")
```

## 9. Visualisation Walk-Forward (Réel vs Prédit)

```python
test_dates = ts.index[-12:]

fig, axes = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('Walk-Forward Validation — Réel vs Prédit', fontsize=14, fontweight='bold')

plot_models = {
    best_hw_name:    best_hw,
    best_sarima_label: best_sarima,
    'Prophet':       prophet_results,
    best_name:       best_results,  # meilleur global mis en avant
}

for ax, (name, res) in zip(axes.flatten(), plot_models.items()):
    ax.plot(ts.index, ts.values, color='#2196F3', lw=1.5, alpha=0.4, label='Historique')
    ax.plot(test_dates, res['actuals'], color='#2196F3', lw=2, marker='o', ms=5, label='Réel')
    ax.plot(test_dates, res['preds'],   color='#FF5722', lw=2, marker='s', ms=5,
            linestyle='--', label=f"Prédit  MAPE={res['MAPE']:.1f}%")
    ax.axvline(test_dates[0], color='gray', linestyle=':', alpha=0.7)
    ax.set_title(name, fontweight='bold')
    ax.legend(fontsize=8)
    ax.grid(alpha=0.3)
    ax.set_ylabel('CA (DT)')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.tick_params(axis='x', rotation=45)
    # Erreur % annotée par mois
    for d, a, p in zip(test_dates, res['actuals'], res['preds']):
        err = abs(a - p) / a * 100
        ax.annotate(f'{err:.0f}%', xy=(d, p), fontsize=6, color='gray', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## 10. Prévision Finale — 12 mois (Fév 2026 → Jan 2027)

```python
print(f"⏳ Entraînement du modèle final : {best_name}\n")

HORIZON = 12

# ── Holt-Winters final ──
def train_final_hw(ts, trend, seasonal):
    model = ExponentialSmoothing(
        ts, trend=trend, seasonal=seasonal,
        seasonal_periods=12, initialization_method='estimated'
    ).fit(optimized=True)
    fc = model.forecast(HORIZON)
    # Intervalles de confiance approx (±1.96 * RMSE)
    rmse_val = best_hw['RMSE']
    lower = fc - 1.96 * rmse_val
    upper = fc + 1.96 * rmse_val
    return fc, lower.clip(lower=0), upper

# ── SARIMA final ──
def train_final_sarima(ts, order, seasonal_order):
    model = SARIMAX(
        ts, order=order, seasonal_order=seasonal_order,
        enforce_stationarity=False, enforce_invertibility=False
    ).fit(disp=False)
    fc_obj = model.get_forecast(steps=HORIZON)
    fc     = fc_obj.predicted_mean
    ci     = fc_obj.conf_int(alpha=0.20)  # IC 80%
    return fc, ci.iloc[:, 0].clip(lower=0), ci.iloc[:, 1]

# ── Prophet final ──
def train_final_prophet(ts):
    df_p = pd.DataFrame({'ds': ts.index, 'y': ts.values})
    model = Prophet(
        seasonality_mode='multiplicative',
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        changepoint_prior_scale=0.05,
        seasonality_prior_scale=10.0,
        interval_width=0.80,
        changepoint_range=0.9,
    )
    model.add_seasonality(name='quarterly', period=91.25, fourier_order=5)
    model.fit(df_p)
    future   = model.make_future_dataframe(periods=HORIZON, freq='MS')
    forecast = model.predict(future).tail(HORIZON)
    fc    = pd.Series(forecast['yhat'].values,       index=pd.date_range(ts.index[-1] + pd.offsets.MonthBegin(1), periods=HORIZON, freq='MS'))
    lower = pd.Series(forecast['yhat_lower'].values, index=fc.index).clip(lower=0)
    upper = pd.Series(forecast['yhat_upper'].values, index=fc.index)
    return fc, lower, upper

# Entraîner le meilleur modèle
if 'Prophet' in best_name:
    forecast_12, lower_12, upper_12 = train_final_prophet(ts)
elif 'SARIMA' in best_name:
    forecast_12, lower_12, upper_12 = train_final_sarima(ts, best_sarima_order, best_sarima_seasonal)
else:
    # Holt-Winters — retrouver les paramètres
    for cfg in hw_configs:
        if cfg['label'] == best_hw_name:
            forecast_12, lower_12, upper_12 = train_final_hw(ts, cfg['trend'], cfg['seasonal'])
            break

# Résumé
print(f"{'Mois':<12} {'Prévision':>14} {'IC Bas (80%)':>14} {'IC Haut (80%)':>14}")
print("-" * 56)
for d, v, lo, hi in zip(forecast_12.index, forecast_12.values, lower_12.values, upper_12.values):
    print(f"  {d.strftime('%Y-%m'):<10} {v:>14,.0f} {lo:>14,.0f} {hi:>14,.0f}")

ca_prevu   = forecast_12.sum()
ca_2025    = ts[ts.index.year == 2025].sum()
ca_2024    = ts[ts.index.year == 2024].sum()
croissance = (ca_prevu - ca_2025) / ca_2025 * 100

print(f"\n{'='*55}")
print(f"📋 CA ANNUEL")
print(f"{'='*55}")
print(f"  CA réel 2024       : {ca_2024:>15,.2f} DT")
print(f"  CA réel 2025       : {ca_2025:>15,.2f} DT")
print(f"  CA prévu 2026      : {ca_prevu:>15,.2f} DT")
print(f"  Croissance prévue  : {croissance:>15.2f} %")
print(f"  Modèle             : {best_name}")
print(f"  MAPE (validation)  : {df_results.iloc[0]['MAPE (%)']:.2f}%")
print(f"{'='*55}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# ── Graphe principal : historique + prévision
ax = axes[0]
ax.plot(ts.index, ts.values, color='#2196F3', lw=2, label='CA Historique')
ax.plot(forecast_12.index, forecast_12.values,
        color='#FF5722', lw=2.5, linestyle='--', marker='o', ms=5,
        label=f'Prévision 12 mois ({best_name})')
ax.fill_between(forecast_12.index, lower_12.values, upper_12.values,
                alpha=0.15, color='#FF5722', label='IC 80%')
ax.axvline(ts.index[-1], color='gray', linestyle=':', alpha=0.7, label="Aujourd'hui")

# Annotations valeurs prévues
for d, v in zip(forecast_12.index, forecast_12.values):
    ax.annotate(f'{v/1000:.0f}k', xy=(d, v), xytext=(0, 8),
                textcoords='offset points', ha='center', fontsize=7, color='#FF5722')

ax.set_title(f'CA Mensuel — Historique & Prévision 12 mois
(Modèle : {best_name}  |  MAPE validation : {df_results.iloc[0]["MAPE (%)"]:.2f}%)',
             fontweight='bold')
ax.set_ylabel('CA (DT)')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.tick_params(axis='x', rotation=45)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# ── Barres CA annuel
ax = axes[1]
years_data = pd.DataFrame([
    {'Année': '2023 (réel)', 'CA': ts[ts.index.year == 2023].sum(), 'color': '#2196F3'},
    {'Année': '2024 (réel)', 'CA': ts[ts.index.year == 2024].sum(), 'color': '#2196F3'},
    {'Année': '2025 (réel)', 'CA': ca_2025,                         'color': '#2196F3'},
    {'Année': '2026 (prévu)', 'CA': ca_prevu,                       'color': '#FF5722'},
])
bars = ax.bar(years_data['Année'], years_data['CA'],
              color=years_data['color'], alpha=0.85, edgecolor='white')
for bar, val in zip(bars, years_data['CA']):
    ax.text(bar.get_x() + bar.get_width()/2, val * 1.01,
            f'{val/1e6:.2f}M DT', ha='center', fontsize=10, fontweight='bold')
ax.set_title('CA Annuel Réel vs Prévu', fontweight='bold')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.grid(alpha=0.3, axis='y')

plt.suptitle('Prévision CA 2026 — Time Series', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 12. Sauvegarde dans ClickHouse `gold.ml_predictions`

```python
# Drop + recréation propre (évite les conflits de schéma)
client.command("DROP TABLE IF EXISTS gold.ml_predictions")
client.command("""
    CREATE TABLE gold.ml_predictions (
        prediction_date  Date,
        year             Int32,
        month            Int32,
        predicted_sales  Float64,
        lower_bound      Float64,
        upper_bound      Float64,
        model_name       String,
        mape_validation  Float64,
        is_prediction    Int8,
        run_timestamp    DateTime DEFAULT now()
    )
    ENGINE = MergeTree()
    ORDER BY (prediction_date, model_name)
""")

# Données à insérer
now = datetime.now()
mape_val = float(df_results.iloc[0]['MAPE (%)'])

data = []
for d, v, lo, hi in zip(forecast_12.index, forecast_12.values, lower_12.values, upper_12.values):
    data.append([
        d.date(),
        int(d.year),
        int(d.month),
        float(v),
        float(lo),
        float(hi),
        best_name,
        mape_val,
        1,
    ])

client.insert(
    'gold.ml_predictions',
    data,
    column_names=['prediction_date','year','month','predicted_sales',
                  'lower_bound','upper_bound','model_name','mape_validation','is_prediction']
)

# Vérification
check = client.query_df('SELECT * FROM gold.ml_predictions ORDER BY prediction_date')
print(f"✅ {len(check)} prévisions écrites dans gold.ml_predictions")
print(f"   Modèle : {best_name}")
print(f"   MAPE   : {mape_val:.2f}%\n")
print(check[['prediction_date','year','month','predicted_sales','lower_bound','upper_bound']].to_string(index=False))
```
