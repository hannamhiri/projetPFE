# 03 - Prédiction des 12 prochains mois
## Objectif : Prédire CA Fév 2026 → Jan 2027
## Variables contemporaines estimées par moyenne des 6 derniers mois

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import joblib
import json
import clickhouse_connect
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.float_format', '{:,.2f}'.format)
print('✅ Imports OK')
```

## 1. Charger modèle + config

```python
model_final  = joblib.load('/app/notebooks/best_model.pkl')
scaler_final = joblib.load('/app/notebooks/scaler.pkl')

with open('/app/notebooks/model_config.json') as f:
    config = json.load(f)

FEATURES_FINAL       = config['features_final']
FEATURES_SELECTED    = config['features_selected']
CONTEMPORAINES       = config['contemporaines']
CONTEMPORAINES_MEANS = config['contemporaines_means']
USE_SCALER           = config['use_scaler']
BEST_MODEL           = config['best_model']
TARGET               = 'total_sales'

df_ml = pd.read_csv('/app/notebooks/ml_dataset.csv', parse_dates=['date'])

print(f'✅ Modèle         : {BEST_MODEL}')
print(f'✅ MAPE test      : {config["mape_test"]:.2f}%')
print(f'✅ MAE test       : {config["mae_test"]:,.0f} DT')
print(f'✅ R² test        : {config["r2_test"]:.4f}')
print(f'✅ Dernier mois   : {df_ml["date"].max().strftime("%b %Y")}')
print(f'\n✅ FEATURES_FINAL ({len(FEATURES_FINAL)}) : {FEATURES_FINAL}')
print(f'\n✅ Moyennes contemporaines :')
for k, v in CONTEMPORAINES_MEANS.items():
    print(f'   {k:<20} : {v:,.2f}')
```

## 2. Prédiction itérative des 12 prochains mois

```python
df_future = df_ml.copy()
last_date = df_future['date'].max()
all_sales = list(df_future[TARGET])

print(f'Départ   : {last_date.strftime("%b %Y")}')
print(f'Objectif : prédire jusqu\'à {(last_date + pd.DateOffset(months=12)).strftime("%b %Y")}\n')

predictions = []

for i in range(1, 13):
    next_date  = last_date + pd.DateOffset(months=i)
    next_month = next_date.month
    next_year  = next_date.year

    row = {}

    # ── Features temporelles ──────────────────────────────
    if 'year'        in FEATURES_FINAL: row['year']        = next_year
    if 'month'       in FEATURES_FINAL: row['month']       = next_month
    if 'quarter'     in FEATURES_FINAL: row['quarter']     = (next_month - 1) // 3 + 1
    if 'semester'    in FEATURES_FINAL: row['semester']    = 1 if next_month <= 6 else 2
    if 'is_summer'   in FEATURES_FINAL: row['is_summer']   = 1 if next_month in [7, 8] else 0
    if 'is_end_year' in FEATURES_FINAL: row['is_end_year'] = 1 if next_month in [11, 12] else 0
    if 'is_january'  in FEATURES_FINAL: row['is_january']  = 1 if next_month == 1 else 0

    # ── Lag features ─────────────────────────────────────
    if 'lag_1'  in FEATURES_FINAL: row['lag_1']  = all_sales[-1]
    if 'lag_2'  in FEATURES_FINAL: row['lag_2']  = all_sales[-2] if len(all_sales) >= 2  else all_sales[-1]
    if 'lag_3'  in FEATURES_FINAL: row['lag_3']  = all_sales[-3] if len(all_sales) >= 3  else all_sales[-1]
    if 'lag_12' in FEATURES_FINAL:
        past_same = df_future[df_future['month'] == next_month][TARGET]
        row['lag_12'] = past_same.iloc[-1] if len(past_same) > 0 else np.mean(all_sales[-12:])

    # ── Rolling features ──────────────────────────────────
    if 'rolling_mean_3' in FEATURES_FINAL: row['rolling_mean_3'] = np.mean(all_sales[-3:])
    if 'rolling_mean_6' in FEATURES_FINAL: row['rolling_mean_6'] = np.mean(all_sales[-6:])

    # ── Variables contemporaines — moyenne 6 derniers mois ─
    for col in CONTEMPORAINES:
        if col in FEATURES_FINAL:
            row[col] = CONTEMPORAINES_MEANS[col]

    # ── Prédiction ────────────────────────────────────────
    X_row = pd.DataFrame([row])[FEATURES_FINAL]
    if USE_SCALER:
        X_row_sc = scaler_final.transform(X_row)
        pred = model_final.predict(X_row_sc)[0]
    else:
        pred = model_final.predict(X_row)[0]

    all_sales.append(pred)
    predictions.append({
        'date':            next_date,
        'year':            next_year,
        'month':           next_month,
        'predicted_sales': pred,
        'is_prediction':   1
    })
    print(f'  {next_date.strftime("%b %Y")} → {pred:>15,.2f} DT')

df_pred = pd.DataFrame(predictions)
print(f'\n✅ {len(df_pred)} mois prédits')
print(f'CA prévisionnel total : {df_pred["predicted_sales"].sum():,.2f} DT')
```

## 3. Visualisation — Réel + Prédit sur un seul graphe

```python
fig = go.Figure()

# CA réel
fig.add_trace(go.Scatter(
    x=df_ml['date'], y=df_ml[TARGET],
    name='CA Réel',
    mode='lines+markers',
    line=dict(color='steelblue', width=2),
    marker=dict(size=5)
))

# CA prédit
fig.add_trace(go.Scatter(
    x=df_pred['date'], y=df_pred['predicted_sales'],
    name=f'CA Prédit ({BEST_MODEL})',
    mode='lines+markers',
    line=dict(color='orange', width=2, dash='dash'),
    marker=dict(size=6, symbol='square')
))

# Ligne de coupure
fig.add_vline(
    x=last_date.timestamp() * 1000,
    line_dash='dot', line_color='red',
    annotation_text=f'{last_date.strftime("%b %Y")} — Fin données réelles'
)

fig.update_layout(
    title=f'📈 CA Réel (Jan 2022 → Jan 2026) + Prédit (Fév 2026 → Jan 2027) — {BEST_MODEL}',
    xaxis_title='Mois',
    yaxis_title='CA (DT)',
    hovermode='x unified',
    height=500
)
fig.show()
```

## 4. CA annuel prévu vs réel 2025

```python
ca_annuel_prevu = df_pred['predicted_sales'].sum()
ca_annuel_2025  = df_ml[df_ml['date'].dt.year == 2025][TARGET].sum()
ca_annuel_2024  = df_ml[df_ml['date'].dt.year == 2024][TARGET].sum()
croissance      = (ca_annuel_prevu - ca_annuel_2025) / ca_annuel_2025 * 100

print('=' * 55)
print('📋 PRÉVISION CA ANNUEL')
print('=' * 55)
print(f'CA réel 2024          : {ca_annuel_2024:>15,.2f} DT')
print(f'CA réel 2025          : {ca_annuel_2025:>15,.2f} DT')
print(f'CA prévu 2026-2027    : {ca_annuel_prevu:>15,.2f} DT')
print(f'Croissance prévue     : {croissance:>15.2f} %')
print(f'Modèle utilisé        : {BEST_MODEL}')
print(f'MAPE sur test         : {config["mape_test"]:.2f}%')
print(f'MAE sur test          : {config["mae_test"]:,.0f} DT')
print('=' * 55)

print('\nDétail mensuel prédictions :')
print(df_pred[['date', 'predicted_sales']].to_string(index=False))

# Graphe annuel
years_data = pd.DataFrame([
    {'Année': '2024 (réel)', 'CA': ca_annuel_2024},
    {'Année': '2025 (réel)', 'CA': ca_annuel_2025},
    {'Année': '2026-27 (prédit)', 'CA': ca_annuel_prevu},
])

fig, ax = plt.subplots(figsize=(8, 5))
colors = ['steelblue', 'steelblue', 'orange']
ax.bar(years_data['Année'], years_data['CA'], color=colors)
for i, (_, row) in enumerate(years_data.iterrows()):
    ax.text(i, row['CA'] + years_data['CA'].max()*0.01,
            f"{row['CA']:,.0f} DT", ha='center', fontsize=9)
ax.set_title('📊 CA Annuel Réel vs Prévu')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.tight_layout()
plt.show()
```

## 5. Écriture dans ClickHouse Gold

```python
client = clickhouse_connect.get_client(
    host='clickhouse', port=8123,
    user='default', password='clickhouse123',
    database='gold'
)

# Créer la table
client.command("""
    CREATE TABLE IF NOT EXISTS gold.ml_predictions (
        date             Date,
        year             Int32,
        month            Int32,
        predicted_sales  Float64,
        is_prediction    Int8
    )
    ENGINE = MergeTree()
    ORDER BY date
""")

# Vider et réinsérer
client.command('TRUNCATE TABLE gold.ml_predictions')
client.insert_df('gold.ml_predictions',
    df_pred[['date', 'year', 'month', 'predicted_sales', 'is_prediction']])

# Vérification
check = client.query_df('SELECT * FROM gold.ml_predictions ORDER BY date')
print(f'✅ {len(check)} prédictions écrites dans ClickHouse Gold')
print('   Table : gold.ml_predictions\n')
print(check.to_string(index=False))
```
