# 03 - Pipeline Cascade + Training
## Objectif : Prédire nb_clients, nb_documents, total_quantity puis total_sales
## ⚠️ Changer PREP_METHOD entre 'corr' et 'modsel' pour comparer les deux

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import joblib
import warnings

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings('ignore')
pd.set_option('display.float_format', '{:,.2f}'.format)
plt.rcParams['figure.figsize'] = (14, 5)
print('✅ Imports OK')
```

## 1. Configuration — Choisir la méthode de preprocessing

```python
# ⚠️ Changer ici pour comparer les deux méthodes
PREP_METHOD = 'corr'   # 'corr' ou 'modsel'

if PREP_METHOD == 'corr':
    DATASET_PATH  = '/app/notebooks/ml_dataset_corr.csv'
    FEATURES_PATH = '/app/notebooks/base_features_corr.json'
else:
    DATASET_PATH  = '/app/notebooks/ml_dataset_modsel.csv'
    FEATURES_PATH = '/app/notebooks/base_features_modsel.json'

TARGET          = 'total_sales'
TARGETS_CASCADE = ['nb_clients', 'nb_documents', 'total_quantity']

df_ml = pd.read_csv(DATASET_PATH, parse_dates=['date'])
with open(FEATURES_PATH) as f:
    BASE_FEATURES = json.load(f)

# S'assurer que BASE_FEATURES existent dans le dataset
BASE_FEATURES = [f for f in BASE_FEATURES if f in df_ml.columns]

print(f'✅ Méthode      : {PREP_METHOD}')
print(f'✅ Dataset      : {df_ml.shape}')
print(f'✅ BASE_FEATURES ({len(BASE_FEATURES)}) : {BASE_FEATURES}')
print(f'✅ Période      : {df_ml["date"].min().strftime("%b %Y")} → {df_ml["date"].max().strftime("%b %Y")}')
```

## 2. Split temporel 85/15

```python
split_idx = int(len(df_ml) * 0.85)

df_train = df_ml.iloc[:split_idx].copy()
df_test  = df_ml.iloc[split_idx:].copy()

print(f'Train : {len(df_train)} mois ({df_train["date"].min().strftime("%b %Y")} → {df_train["date"].max().strftime("%b %Y")})')
print(f'Test  : {len(df_test)} mois ({df_test["date"].min().strftime("%b %Y")} → {df_test["date"].max().strftime("%b %Y")})')

fig, ax = plt.subplots()
ax.plot(df_train['date'], df_train[TARGET], label='Train', color='steelblue', marker='o')
ax.plot(df_test['date'],  df_test[TARGET],  label='Test',  color='orange',    marker='o')
ax.axvline(df_train['date'].max(), color='red', linestyle='--', label='Split')
ax.set_title('📊 Split Temporel Train / Test')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
plt.tight_layout()
plt.show()
```

## 3. Entraîner les modèles intermédiaires (cascade)

```python
cascade_models  = {}
cascade_scalers = {}

for target_var in TARGETS_CASCADE:
    print(f'\n{"="*45}')
    print(f'🎯 Modèle intermédiaire : {target_var}')
    print(f'{"="*45}')

    X_tr = df_train[BASE_FEATURES]
    y_tr = df_train[target_var]
    X_te = df_test[BASE_FEATURES]
    y_te = df_test[target_var]

    sc      = StandardScaler()
    X_tr_sc = sc.fit_transform(X_tr)
    X_te_sc = sc.transform(X_te)

    m = Ridge(alpha=10)
    m.fit(X_tr_sc, y_tr)
    pred = m.predict(X_te_sc)

    mae  = mean_absolute_error(y_te, pred)
    mape = np.mean(np.abs((y_te - pred) / y_te)) * 100
    r2   = r2_score(y_te, pred)

    cascade_models[target_var]  = m
    cascade_scalers[target_var] = sc

    print(f'MAE  : {mae:>12,.2f}')
    print(f'MAPE : {mape:>12.2f} %')
    print(f'R²   : {r2:>12.4f}')

print(f'\n✅ {len(cascade_models)} modèles intermédiaires entraînés')
```

## 4. Augmenter train et test avec les prédictions intermédiaires

```python
df_train_aug = df_train.copy()
df_test_aug  = df_test.copy()

for target_var in TARGETS_CASCADE:
    m  = cascade_models[target_var]
    sc = cascade_scalers[target_var]

    X_tr_sc = sc.transform(df_train[BASE_FEATURES])
    df_train_aug[f'{target_var}_pred'] = m.predict(X_tr_sc)

    X_te_sc = sc.transform(df_test[BASE_FEATURES])
    df_test_aug[f'{target_var}_pred'] = m.predict(X_te_sc)

PRED_VARS     = [f'{v}_pred' for v in TARGETS_CASCADE]
FINAL_FEATURES = BASE_FEATURES + PRED_VARS

print(f'✅ Features finales ({len(FINAL_FEATURES)}) :')
for f in FINAL_FEATURES:
    print(f'  • {f}')
```

## 5. Comparer les modèles finaux sur total_sales

```python
X_train = df_train_aug[FINAL_FEATURES]
X_test  = df_test_aug[FINAL_FEATURES]
y_train = df_train_aug[TARGET]
y_test  = df_test_aug[TARGET]

scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

def evaluate(name, model, X_tr, y_tr, X_te, y_te):
    model.fit(X_tr, y_tr)
    pred  = model.predict(X_te)
    mae   = mean_absolute_error(y_te, pred)
    rmse  = np.sqrt(mean_squared_error(y_te, pred))
    r2    = r2_score(y_te, pred)
    mape  = np.mean(np.abs((y_te - pred) / y_te)) * 100
    smape = np.mean(2 * np.abs(y_te - pred) / (np.abs(y_te) + np.abs(pred))) * 100
    bias  = np.mean(pred - y_te)
    return {'name': name, 'model': model, 'pred': pred,
            'MAE': mae, 'RMSE': rmse, 'R²': r2,
            'MAPE': mape, 'SMAPE': smape, 'Biais': bias}

models_config = [
    ('LinearRegression', LinearRegression(),                                          X_train_sc, X_test_sc),
    ('Ridge',            Ridge(alpha=10),                                             X_train_sc, X_test_sc),
    ('DecisionTree',     DecisionTreeRegressor(max_depth=3, random_state=42),        X_train,    X_test),
    ('RandomForest',     RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42), X_train, X_test),
    ('XGBoost',          XGBRegressor(n_estimators=300, max_depth=3, learning_rate=0.03, random_state=42), X_train, X_test),
]

results = []
for name, model, Xtr, Xte in models_config:
    r = evaluate(name, model, Xtr, y_train, Xte, y_test)
    results.append(r)

df_results = pd.DataFrame(results).set_index('name')

print('=' * 80)
print(f'{"Modèle":<20} {"MAE":>12} {"RMSE":>12} {"R²":>8} {"MAPE":>8} {"SMAPE":>8} {"Biais":>12}')
print('=' * 80)
for _, row in df_results.iterrows():
    print(f'{row.name:<20} {row["MAE"]:>12,.0f} {row["RMSE"]:>12,.0f} '
          f'{row["R²"]:>8.4f} {row["MAPE"]:>7.2f}% {row["SMAPE"]:>7.2f}% {row["Biais"]:>12,.0f}')
print('=' * 80)

best_name = df_results['MAPE'].idxmin()
print(f'\n🏆 Meilleur modèle : {best_name} (MAPE = {df_results.loc[best_name, "MAPE"]:.2f}%)')
```

## 6. Détail mois par mois — meilleur modèle

```python
best = df_results.loc[best_name]

df_detail = df_test[['date']].copy()
df_detail['reel']   = y_test.values
df_detail['predit'] = best['pred']
df_detail['ecart']  = best['pred'] - y_test.values
df_detail['ecart%'] = (df_detail['ecart'] / df_detail['reel'] * 100).round(2)

print(f'=== Détail {best_name} ({PREP_METHOD}) ===\n')
print(df_detail.to_string(index=False))
print(f"\nBiais : {best['Biais']:,.2f} DT  ({'surestimation' if best['Biais'] > 0 else 'sous-estimation'})")

fig, ax = plt.subplots()
ax.plot(df_test['date'], y_test.values,
        label='CA Réel', color='steelblue', marker='o', linewidth=2)
ax.plot(df_test['date'], best['pred'],
        label=f'CA Prédit {best_name} (MAPE={best["MAPE"]:.1f}%)',
        color='orange', marker='s', linestyle='--', linewidth=2)
ax.fill_between(df_test['date'], y_test.values, best['pred'], alpha=0.15, color='red')
ax.set_title(f'📈 CA Réel vs Prédit — {best_name} ({PREP_METHOD})')
ax.set_ylabel('CA (DT)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
plt.tight_layout()
plt.show()
```

## 7. Réentraîner sur tout le dataset + Sauvegarde

```python
print(f'Réentraînement sur {len(df_ml)} mois complets...')

# Augmenter tout le dataset avec les prédictions cascade
df_full_aug = df_ml.copy()
for target_var in TARGETS_CASCADE:
    m  = cascade_models[target_var]
    sc = cascade_scalers[target_var]
    X_full_sc = sc.transform(df_ml[BASE_FEATURES])
    df_full_aug[f'{target_var}_pred'] = m.predict(X_full_sc)

X_full = df_full_aug[FINAL_FEATURES]
y_full = df_full_aug[TARGET]

# Réentraîner le meilleur modèle sur tout
scaler_final = StandardScaler()
best_model_obj = df_results.loc[best_name, 'model']
use_scaler = best_name in ['LinearRegression', 'Ridge']

if use_scaler:
    X_full_sc    = scaler_final.fit_transform(X_full)
    model_final  = type(best_model_obj)(**best_model_obj.get_params())
    model_final.fit(X_full_sc, y_full)
else:
    model_final = type(best_model_obj)(**best_model_obj.get_params())
    model_final.fit(X_full, y_full)
    scaler_final = scaler

# Réentraîner aussi les modèles cascade sur tout
cascade_models_final  = {}
cascade_scalers_final = {}
for target_var in TARGETS_CASCADE:
    sc_f = StandardScaler()
    X_full_sc_f = sc_f.fit_transform(df_ml[BASE_FEATURES])
    m_f = Ridge(alpha=10)
    m_f.fit(X_full_sc_f, df_ml[target_var])
    cascade_models_final[target_var]  = m_f
    cascade_scalers_final[target_var] = sc_f

# Sauvegarder
joblib.dump(model_final,           '/app/notebooks/best_model.pkl')
joblib.dump(scaler_final,          '/app/notebooks/scaler.pkl')
joblib.dump(cascade_models_final,  '/app/notebooks/cascade_models.pkl')
joblib.dump(cascade_scalers_final, '/app/notebooks/cascade_scalers.pkl')

with open('/app/notebooks/cascade_config.json', 'w') as f:
    json.dump({
        'best_model':       best_name,
        'prep_method':      PREP_METHOD,
        'base_features':    BASE_FEATURES,
        'targets_cascade':  TARGETS_CASCADE,
        'pred_vars':        PRED_VARS,
        'final_features':   FINAL_FEATURES,
        'use_scaler':       use_scaler,
        'mape_test':        float(best['MAPE']),
        'mae_test':         float(best['MAE']),
    }, f)

print(f'✅ Pipeline cascade sauvegardé')
print(f'   Modèle final    : {best_name}')
print(f'   Méthode prep    : {PREP_METHOD}')
print(f'   MAPE test       : {best["MAPE"]:.2f}%')
print(f'   MAE test        : {best["MAE"]:,.0f} DT')
print(f'   Entraîné sur    : {len(df_ml)} mois')
```
