# 02 - Feature Engineering
## Objectif : Préparer les features pour le modèle de prévision des ventes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import clickhouse_connect
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.float_format', '{:,.2f}'.format)
plt.rcParams['figure.figsize'] = (14, 5)
sns.set_theme(style='whitegrid')

print("✅ Imports OK")
```

```python
client = clickhouse_connect.get_client(
    host='clickhouse',
    port=8123,
    user='default',
    password='clickhouse123',
    database='gold'
)

query = """
SELECT
    d.Year       AS year,
    d.Month      AS month,
    d.Quarter    AS quarter,
    SUM(f.LinePrice)               AS total_sales,
    SUM(f.Quantity)                AS total_quantity,
    COUNT(DISTINCT f.DocumentCode) AS nb_documents,
    COUNT(DISTINCT f.ClientSK)     AS nb_clients
FROM gold.fact_sales AS f
JOIN gold.dim_date AS d ON f.DateSK = d.DateSK
WHERE f.DocumentTypeSK IN (4, 1, 5)
  AND f.LinePrice IS NOT NULL
GROUP BY year, month, quarter
ORDER BY year, month
"""

df = client.query_df(query)
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

print(f"✅ {len(df)} mois chargés")
df.head()
```

## 1. Nettoyage — Suppression du mois incomplet

```python
# Suppression des mois incomplets (début et fin de période)
print(f"Avant nettoyage : {len(df)} mois")
print(f"\nMois incomplets détectés :")
print(df[
    (df['date'] < '2022-01-01') | (df['date'] >= '2026-02-01')
][['date', 'total_sales']].to_string(index=False))

df = df[
    (df['date'] >= '2022-01-01') & (df['date'] < '2026-02-01')
].reset_index(drop=True)

print(f"\nAprès nettoyage : {len(df)} mois")
print(f"Période finale  : {df['date'].min().strftime('%b %Y')} → {df['date'].max().strftime('%b %Y')}")
print(f"Dernier mois    : {df.iloc[-1]['date'].strftime('%b %Y')} → {df.iloc[-1]['total_sales']:,.2f} DT")
```

## 2. Features temporelles

```python
# Déjà disponibles : year, month, quarter
# On ajoute des transformations cycliques pour capturer la saisonnalité

# Encodage cyclique du mois (sin/cos) — meilleur que le mois brut pour ML
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

# Tendance linéaire (index temporel)
df['trend'] = np.arange(len(df))

# Semestre
df['semester'] = df['month'].apply(lambda x: 1 if x <= 6 else 2)

print("✅ Features temporelles créées")
print(df[['date', 'month', 'month_sin', 'month_cos', 'trend', 'semester']].head(6))
```

## 3. Lag features — valeurs passées

```python
# Le modèle apprend à partir des CA des mois précédents
df['lag_1']  = df['total_sales'].shift(1)   # CA mois-1
df['lag_2']  = df['total_sales'].shift(2)   # CA mois-2
df['lag_3']  = df['total_sales'].shift(3)   # CA mois-3
df['lag_6']  = df['total_sales'].shift(6)   # CA il y a 6 mois
df['lag_12'] = df['total_sales'].shift(12)  # CA même mois année précédente

print("✅ Lag features créées")
print(df[['date', 'total_sales', 'lag_1', 'lag_2', 'lag_3', 'lag_6', 'lag_12']].head(15))
```

## 4. Rolling features — moyennes glissantes

```python
# Moyennes glissantes — lissent le bruit et capturent la tendance
df['rolling_mean_3']  = df['total_sales'].shift(1).rolling(window=3).mean()
df['rolling_mean_6']  = df['total_sales'].shift(1).rolling(window=6).mean()
df['rolling_std_3']   = df['total_sales'].shift(1).rolling(window=3).std()

# Ratio CA vs moyenne glissante 6 mois → détecte les pics/creux
df['ratio_vs_rolling6'] = df['lag_1'] / df['rolling_mean_6']

print("✅ Rolling features créées")
print(df[['date', 'total_sales', 'rolling_mean_3', 'rolling_mean_6', 
          'rolling_std_3', 'ratio_vs_rolling6']].head(15))
```

## 5. Features utiles pour les pièces auto

```python
# À la place de is_ramadan, ces events ont plus de sens :

# 1. Été (Juil/Août) → plus de voyages → plus d'accidents/usure
df['is_summer'] = df['month'].isin([7, 8]).astype(int)

# 2. Fin d'année (Nov/Déc) → révisions avant hiver
df['is_end_year'] = df['month'].isin([11, 12]).astype(int)

# 3. Début d'année (Jan) → creux traditionnel
df['is_january'] = (df['month'] == 1).astype(int)
```

## 6. Suppression des lignes avec NaN (dues aux lags)

```python
print(f"Avant dropna : {len(df)} mois")
print(f"NaN par colonne :\n{df.isnull().sum()[df.isnull().sum() > 0]}\n")

df_ml = df.dropna().reset_index(drop=True)

print(f"Après dropna  : {len(df_ml)} mois")
print(f"Période finale : {df_ml['date'].min().strftime('%b %Y')} → {df_ml['date'].max().strftime('%b %Y')}")
```

## 7. Visualisation des features

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

lag_cols = ['lag_1', 'lag_2', 'lag_3', 'lag_6', 'lag_12', 'rolling_mean_6']

for i, col in enumerate(lag_cols):
    axes[i].scatter(df_ml[col], df_ml['total_sales'], alpha=0.6, color='steelblue')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('total_sales')
    axes[i].set_title(f'total_sales vs {col}')
    
    # Corrélation
    corr = df_ml[col].corr(df_ml['total_sales'])
    axes[i].set_title(f'{col}  (corr={corr:.2f})')

plt.suptitle('📊 Corrélation Features Lag vs CA Réel', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
feature_cols = ['total_sales', 'lag_1', 'lag_2', 'lag_3', 'lag_6', 'lag_12',
                'rolling_mean_3', 'rolling_mean_6', 'month_sin', 'month_cos',
                'trend', 'quarter', 'semester',
                'is_summer', 'is_end_year', 'is_january']

corr = df_ml[feature_cols].corr()

fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, ax=ax, linewidths=0.5)
ax.set_title('🔥 Corrélation entre toutes les features')
plt.tight_layout()
plt.show()
```

## 8. Dataset final

```python
FEATURES = [
    'lag_1', 'lag_2', 'lag_3', 'lag_6', 'lag_12',
    'rolling_mean_3', 'rolling_mean_6', 'rolling_std_3',
    'ratio_vs_rolling6', 'month_sin', 'month_cos',
    'trend', 'quarter', 'semester',
    'is_summer',       # ← nouveau
    'is_end_year',     # ← nouveau
    'is_january',      # ← nouveau
    # 'is_ramadan'     # ← supprimé
]
TARGET = 'total_sales'

X = df_ml[FEATURES]
y = df_ml[TARGET]

print("=" * 50)
print("📋 DATASET FINAL POUR ML")
print("=" * 50)
print(f"Nb échantillons  : {len(df_ml)}")
print(f"Nb features      : {len(FEATURES)}")
print(f"Target           : {TARGET}")
print(f"Période          : {df_ml['date'].min().strftime('%b %Y')} → {df_ml['date'].max().strftime('%b %Y')}")
print(f"\nFeatures :\n")
for f in FEATURES:
    print(f"  • {f}")
print("=" * 50)

df_ml[['date', TARGET] + FEATURES].tail(10)
```

```python
df_ml.to_csv('/app/notebooks/ml_dataset.csv', index=False)
print(f"✅ Sauvegardé → {len(df_ml)} lignes, {len(FEATURES)} features")
```
